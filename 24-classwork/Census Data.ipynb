# Exploring columns

Yes, we can select one column at a time, as always. Check the `acs-2017-data-dictionary.txt` file to understand the meaning of the columns. **You'll need to add `SE_` to them for them to work as column names.**

```
A12001_001:     Population 25 Years and Over:
```

````python
divisions_2017['SE_A07001_001']
````

Since we set the index, graphing is really easy and nice.

````python
divisions_2017['SE_A07001_001'].plot(kind='bar')
````

If we want to move that into [DataWrapper](https://www.datawrapper.de/), we just copy it to the clipboard and paste it in.

````python
divisions_2017['SE_A07001_001'].to_clipboard(header=True)
````

A few button-presses later and [there we go](https://datawrapper.dwcdn.net/loPmU/1/)!

When you click "grouped bar chart," the graph you're looking at **doesn't look like the one above**. Click the "transpose" button and it will sort itself out and give you [this similar graphic](https://datawrapper.dwcdn.net/b5AMJ/1/).

# Joining multiple years together

First we'll think about what we want. **Remember, we need to add `SE_` to our column names.**

```
      A09003_001:     Average Commute to Work (In Min)
```

````python
national_2017['SE_A09003_001']
````

````python
national_2011['SE_A09003_001']
````

````python
merged = pd.merge(national_2011['SE_A09003_001'],
                  national_2017['SE_A09003_001'],
                  left_index=True,
                  right_index=True,
                  suffixes=['_2011', '_2017'])
merged
````

````python
merged.to_clipboard()
````

DataWrapper yells that we don't have enough data, but after a quick click of the "transpose" button [we're all set](https://datawrapper.dwcdn.net/riHYC/1/)

## Joining multiple years and multiple columns

Multiple years and multiple columns becomes a real chore, because DataWrapper doesn't like small multiples. Depending on your chart, you'll probably do something like this, though:

````python
education_2011 = divisions_2011.loc[:, 'SE_A12001_002':'SE_A12001_008']
education_2011['year'] = 2011

education_2017 = divisions_2017.loc[:, 'SE_A12001_002':'SE_A12001_008']
education_2017['year'] = 2017

combined = pd.concat([education_2011, education_2017]).reset_index()
combined
````

````python
combined.to_clipboard()
````
