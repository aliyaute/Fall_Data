# Powerplants

```python
powerplants = pd.read_csv("powerplants.csv")
powerplants.head(2)
```

```python
# How many of each type
powerplants.groupby(['StateName', 'PrimSource']).Total_MW.count().unstack().fillna(0)
```

```python
# How much power from each type
powerplants.groupby(['StateName', 'PrimSource']).Total_MW.sum().unstack().fillna(0)
```

```python
# How much of one or two or three types
powerplants.groupby(['StateName', 'PrimSource']).Total_MW.sum().unstack().fillna(0)[['nuclear', 'coal']]
```
